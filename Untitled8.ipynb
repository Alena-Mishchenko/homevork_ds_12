```python
class_names = ['Футболка/топ', 'Штани', 'Светр', 'Сукня', 'Пальто', 'Сандалі', 'Сорочка', 'Кросівки', 'Сумка', 'Чоботи']
```

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt

# Завантаження Fashion MNIST
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()


print("Розмір тренувальної вибірки:", x_train.shape)
print("Розмір тестової вибірки:", x_test.shape)


print("Унікальні мітки:", np.unique(y_train))
```

```python
plt.figure(figsize=(10, 10))
for i in range(25):
    plt.subplot(5, 5, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_train[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[y_train[i]])
plt.show()
```

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
y_train = keras.utils.to_categorical(y_train, num_classes=10)
y_test = keras.utils.to_categorical(y_test, num_classes=10)
```

```python
def reinit_model():
    model = keras.Sequential([
        keras.layers.Flatten(input_shape=(28, 28)),
        keras.layers.Dense(128, activation='relu'),
        keras.layers.Dense(10, activation='softmax')
    ])

    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model
```

```python
model = reinit_model()
```

```python
from tensorflow.keras import regularizers
```

```python
!pip install scikeras
from sklearn.model_selection import GridSearchCV
from scikeras.wrappers import KerasClassifier
```

```python
def create_model_with_hparams(
    hidden_layers: int = 2,
    hidden_units: int = 128,
    hidden_activation: str = "relu",
    kernel_initializer: str = "he_normal",
    use_l2: bool = False, use_dropout: bool = False,
    optimizer: str = "adam",
):
    model = keras.Sequential()
    model.add(keras.layers.Flatten(input_shape=(28, 28)))

    for _ in range(hidden_layers):
        model.add(keras.layers.Dense(
            hidden_units,
            activation=hidden_activation,
            kernel_initializer=kernel_initializer,
            kernel_regularizer=(regularizers.l2() if use_l2 else None)
        ))
        if use_dropout:
            model.add(keras.layers.Dropout(0.5))
    model.add(keras.layers.Dense(10, activation='softmax'))

    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model
```

```python
# Створення об'єкту KerasClassifier
model = KerasClassifier(model=create_model_with_hparams, optimizer="adam", hidden_layers=2)
model.get_params().keys()
```

```python

param_grid = {
  #  'model__optimizer': ['adam', 'rmsprop'],
    'model__hidden_layers': [2, 3],
    'model__hidden_units': [32, 64],
    # 'model__hidden_activation': ['relu', 'tanh'],
    # 'model__kernel_initializer': ['he_normal', 'glorot_uniform'],
    'model__use_l2': [ False], # Changed 'use_l2' to 'model__use_l2'
    'model__use_dropout': [True, False],
    'batch_size': [32, 64],
    'epochs': [10]
}


grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=3, verbose=2)

grid_result = grid.fit(x_train, y_train)
```

```python
print(f"Найкращі параметри: {grid_result.best_params_}")
print(f"Найкращий результат: {grid_result.best_score_}")
```
